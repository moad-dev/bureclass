## Дообучение and testing Sentence BERT

````python
import pandas as pd
train_dataset = pd.read_csv("train.csv")
train_dataset.columns=["x", "y"]
train_dataset
````

````python
%pip install scikit-learn --quiet
````

````python
from sklearn.model_selection import train_test_split
X_train, X_test = train_test_split(train_dataset, test_size=0.25, random_state=42)
X_train
````

````python
opendata = pd.read_csv("openchat_gen.csv", sep=";").dropna()
opendata
````

````python
train_dataset = pd.concat([X_train, opendata])
train_dataset = train_dataset.dropna()
train_dataset
````

````python
import math
from sentence_transformers import SentenceTransformer, InputExample, losses
from torch.utils.data import DataLoader

train_set = []
for index, row in train_dataset.iterrows():
    train_set.append(InputExample(texts=[row['x'], row['y']]))

finetuned_model = SentenceTransformer("multilingual-e5-large-amethyst", device="cuda")

train_dataloader = DataLoader(train_set, shuffle=True, batch_size=16)
train_loss = losses.MultipleNegativesRankingLoss(finetuned_model)
# train_loss = losses.MegaBatchMarginLoss(finetuned_model)

num_epochs = 5
warmup_steps = math.ceil(len(train_set) * num_epochs * 0.1)

finetuned_model.fit(train_objectives=[(train_dataloader, train_loss)],
                    epochs=num_epochs,
                    warmup_steps=warmup_steps,
                    output_path="multilingual-e5-large-amethyst-advanced")
````

## Векторный индекс документов

### Структура БД

````python
from typing import Optional
from pgvector.sqlalchemy import Vector
from sqlalchemy import Text, create_engine, select, text
from sqlalchemy.orm import DeclarativeBase, Mapped, Session, mapped_column

engine = create_engine(f"postgresql://{environ.get('POSTGRES_USER')}:{environ.get('POSTGRES_PASSWORD')}@{environ.get('POSTGRES_HOST')}/{environ.get('POSTGRES_DB')}", echo=False)
````

````python
class Base(DeclarativeBase):
    pass

class BuRe(Base):
    __tablename__ = "bure"
    id: Mapped[int] = mapped_column(primary_key=True)
    text: Mapped[str] = mapped_column(Text())
    embedding: Mapped[Optional[Vector]] = mapped_column(Vector(1024))
````

````python
with Session(engine) as session:
    session.execute(text('CREATE EXTENSION IF NOT EXISTS vector'))
    session.execute(text('DROP TABLE bure'))
    session.commit()
Base.metadata.create_all(engine)
````

### Выбор нужного фрагмента через векторный индекс

````python
embeddings_df = pd.read_pickle("ksr_embeddings_advanced.pkl")
embeddings_df
````

````python
with Session(engine) as session:
    for row, document in embeddings_df.iterrows():
        doc = BuRe(
            text=document["name"],
            embedding=document["embedding"],
        )
        session.add(doc)
    session.commit()
````

````python
from sentence_transformers import SentenceTransformer
finetuned_model = SentenceTransformer('saved_models/multilingual-e5-large-amethyst-advanced', device="cpu")
finetuned_model
````

````python
final_test = pd.read_csv("final_test.csv", index_col=0)
final_test
````

````python
def get_nearest_text(text):
    with Session(engine) as session:
        return session.scalars(select(BuRe)
                        .order_by(BuRe.embedding.cosine_distance(
                            finetuned_model.encode(text)
                            )).limit(1)).first().text

get_nearest_text("Комплект заделки стыков КЗС (Т) для предизолир...	")
````

````python
%%timeit
get_nearest_text("Комплект для изоляции стыкового соединения (муфта, пенопакет, термоклей, дренажные и вварные пробки), тип КИС-110, длина 600 мм")
````

````python
final_test["y_pred"] = final_test["x"].apply(get_nearest_text)
final_test.head(5)
````
