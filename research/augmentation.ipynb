## Подготовка датасетов

````python
big_ksr = pd.read_csv("ksr.csv")
big_ksr
````

````python
big_ksr = big_ksr[((big_ksr["type"] == "группа") | (big_ksr["type"] == "ресурс"))]
big_ksr['full_code'] = big_ksr['okpd2'].apply(lambda x: "" if x is np.NaN else x) + "." + big_ksr['code']
big_ksr['full_code'] = big_ksr['full_code'].apply(lambda x: x[1:] if x[0] == '.' else x)
big_ksr = big_ksr[['full_code', 'name', 'type', 'unit']]
big_ksr = big_ksr.reset_index(drop=True)
big_ksr
````

````python
groups = big_ksr[big_ksr['type'] == 'группа']
groups
````

````python
def get_ksr_group(full_code):
  cut_code = full_code.split('-')[0]
  group_code_numbers = cut_code.split('.')[4:]
  return ".".join(group_code_numbers)

get_ksr_group("19.20.42.121.01.2.03.01-1008-000")
````

````python
def get_ksr_name(code):
  return big_ksr[big_ksr['full_code'] == code]['name'].values[0]

get_ksr_name("19.20.42.121.01.2.03.01-1008-000")
````

````python
def get_ksr_code(name):
  return big_ksr[big_ksr['name'] == name]['full_code'].values[0]

get_ksr_code("Вяжущее полимерно-битумное дорожное ПБВ 40")
````

````python
def get_group_list(group_code):
  group_list = []
  group_index = big_ksr[big_ksr["full_code"] == group_code].index.values[0]
  for i in range(group_index+1, len(big_ksr)):
    row = big_ksr.iloc[i]
    if row['type'] == "ресурс":
      group_list.append(row['name'])
    if row['type'] == "группа":
      break
  return group_list

get_group_list("01.2.03.01")
````

## Генерация

````python
from os import environ
from dotenv import load_dotenv
load_dotenv(dotenv_path="../.env")

llm_host = environ.get('LLM_HOST')
````

````python
from langchain.prompts import PromptTemplate

prompt_template = """
Действуй как инженер-строитель, пытающийся закупить у поставщика строительные ресурсы (материалы или машины) из заданной группы ресурсов. Ты пытаешься своими словами описать поставщику смету нужных ресурсов из заданной группы.  

Требуется привести 5 содержательных примеров нужных тебе ресурсов из заданной группы, используя понятный поставщику строительных ресурсов язык. Используй аббревиатуры, опечатки, единицы измерения и количество нужных ресурсов.

Например для группы \"Краны газовые\": 
\"ППР Кран шаровой  Ø 20 мм\"
\"Стальной шаровый кран Ду25, PN40\"
\"Кран шаровый полипропиленовый  д 20\"
\"Кран шаровый стальной, фланце вый., ст. пр., ручка, Ду50, Тмакс. = 180С, Py40 Гросс КШСФ-СП5040Р  689819 Гросс\"
\"50 Полипропиленовый шаровой кран  полнопроходной (PP-R)\"

Сгенерируй для группы \"{content}\" 5 названий строительных ресурсов (материалов или машин). При этом используй аббревиатуры, переставляй слова, добавляй какие-то обозначения и ни в коем случае не повторяйся. Между названиями поставь ```\n```. Группу не упомянай
"""

prompt = PromptTemplate.from_template(prompt_template)
````

````python
from langchain_openai import ChatOpenAI

openchat = ChatOpenAI(
    model="saiga_llama3_8b",
    openai_api_key="EMPTY",
    openai_api_base=llm_host,
    temperature=0.7,
)

openchat_chain = prompt | openchat
openchat_chain
````
